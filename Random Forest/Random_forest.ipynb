```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
x=x.drop(["customerID"],axis=1)
```

```python
x
```

```python
y=df["Churn"]
```

```python
y
```

```python
# Splitting into train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=45)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=50,random_state=45)
```

```python
#training model
model.fit(x_train,y_train)
```

```python
# model.predict(x_train) prediction for train data
```

```python
# Prediction for test data
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_pred,y_test)
```

```python
li=[50,100,150,200,250,300,350,400,450,500,550,600,650,700,750,800,850,900,950,1000]
```

```python
for i in li:
    mo=RandomForestClassifier(n_estimators=i,random_state=56)
    mo.fit(x_train,y_train)
    model_pred=mo.predict(x_test)
    acc =accuracy_score(y_test,model_pred)*100
    print("no. of trees -",i,end=",")
    print("Accuracy",acc)    
```

```python
rf_model=RandomForestClassifier(n_estimators=40,random_state=56,criterion="entropy")
rf_model.fit(x_train,y_train)
rf_model_pred=rf_model.predict(x_test)
acc =accuracy_score(y_test,rf_model_pred)*100
```

```python
acc
```

```python
sns.heatmap(confusion_matrix(rf_model_pred,y_test),annot=True,fmt="G")
```

```python
print(classification_report(rf_model_pred,y_test))
```
