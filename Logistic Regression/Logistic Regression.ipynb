```python
out_columns=["age","campaign","duration","cons.conf.idx"]
```

```python
df["pdays"].value_counts()
```

```python
for col in out_columns:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    df=df[(df[col]>=lower)&(df[col]<=upper)]
```

```python
#outlier Detection
for col in out_columns:
    if df[col].dtype!="object":
        sns.boxplot(df[col])
        plt.xlabel(col)
        plt.ylabel('count')
        plt.show()
```
